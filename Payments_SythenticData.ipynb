# Generate 96 week days at random between Jan 1,2022 and Dec 31st, 2022

```python
import random
import datetime
import pandas as pd

# list to store the dates
dates = []

# start date and end date
start_date = datetime.datetime(2022, 1, 1)
end_date = datetime.datetime(2022, 12, 31)

# loop through all dates between start_date and end_date
current_date = start_date
while current_date <= end_date:
    # check if the current date is a weekday
    if current_date.weekday() < 5:
        dates.append(current_date.strftime("%m/%d"))

    # move to the next day
    current_date += datetime.timedelta(days=1)

# create a dictionary to store the dates for each month
monthly_dates = {}

# loop through the dates and add 8 random dates for each month to the dictionary
for date in dates:
    month = int(date.split('/')[0])
    if month not in monthly_dates:
        monthly_dates[month] = []
    if len(monthly_dates[month]) < 8:
        monthly_dates[month].append(date)

# convert the dictionary to a dataframe
df = pd.DataFrame.from_dict(monthly_dates, orient='index')

# write the dataframe to an excel file
df.to_excel('Processing Dates.xlsx', index=False)
```
